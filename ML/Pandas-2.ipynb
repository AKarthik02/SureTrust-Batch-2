#### Restructuring the Data

```python
weather = pd.read_csv('Datasets/weather.csv')
weather.head()
```

```python
## days columns into single column. 
# Melt operation -> Columns into Rows
```

```python
pd.melt?
```

```python
a = pd.melt(weather,
          id_vars = ['year', 'month', 'element'],
          var_name = "day" ,
        value_name = 'temp')
a.head(10)
```

```python
weather.shape
```

```python
a.shape
```

```python
## max and min as separate columns. For rows into columns we use pivot. 
```

```python
pd.pivot_table?
```

```python
b = a.pivot_table(index = ['year', 'month', 'day'],
                 columns = 'element',
                 values = 'temp').reset_index()
b
```

```python
b.to_csv('Datasets/weather_restructured.csv', index=False)
```

#### Quering

```python
b.loc[(b.month==3) & (b.day=='day1')]
```

####  Custom Methods

```python
def mean_custom(x):
    return sum(x)/len(x)
```

```python
b[['min', 'max']].apply(mean_custom, axis = 1)
```
